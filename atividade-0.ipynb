```python
#importando a biblioteca matplotlib
import matplotlib.pyplot as plt
```

```python
# Configura√ß√£o do plano cartesiano
plt.figure(figsize=(6, 6),)
plt.axhline(y=0, color='white', linewidth=1)  # Eixo X
plt.axvline(x=0, color='white', linewidth=1)  # Eixo Y

# Limites dos eixos
plt.xlim(-8, 8)
plt.ylim(-8, 8)

# Grade
plt.grid(True, alpha=1)

# Get current axes
ax = plt.gca()  

# Cor de fundo do gr√°fico 
ax.set_facecolor('black')  

plt.grid(True, color='white', alpha=0.3, linestyle='-')

# T√≠tulos
plt.title('Plano Cartesiano')

plt.show()
```
