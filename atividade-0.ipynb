##### A transformação linear de rotação gira um vetor ao redor de um ponto fixo, num valor de ângulo específico, no caso da atividade em 90° horários. A matriz de rotação pode ser obtida seguindo duas lógicas distintas, a primeira é indo por lógica e verificando com vetores unitários o que ocorre quando eles são rotacionados, a partir deles pode ser montada uma matriz. A outra forma é por meio de formulação trigonométrica usando senθ e cosθ, mas que resulta na mesma matriz formada por um processo mais simples.

### Rotação horária em 90°:
Seguindo pela lógica menos complicada, utilizando dois vetores unitários, da base canônica em R^2, sendo eles: f=[1,0] e g=[0,1] 

```python
plt.figure(figsize=(8, 8))

# Plano cartesiano
plt.axhline(y=0, color='white', linewidth=1, alpha=0.5)
plt.axvline(x=0, color='white', linewidth=1, alpha=0.5)

# Grade
plt.grid(True, alpha=0.2)

# Vetores da base canônica
plt.quiver(0, 0, 1, 0, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='f = (1,0)')
plt.quiver(0, 0, 0, 1, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='g = (0,1)')

# Pontos finais
plt.scatter([1, 0], [0, 1], color=['red', 'blue'], s=50, zorder=5)

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend(fontsize=12)
plt.title('Base Canônica de ℝ²', fontsize=14)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

ax = plt.gca()  
ax.set_facecolor('black')  

plt.show()
```

Se aplica a rotação de 90°, transformando esses vetores em: [0,-1] e [1,0], resultando nos vetores abaixo:

```python
plt.figure(figsize=(8, 8))

# Plano cartesiano
plt.axhline(y=0, color='white', linewidth=1, alpha=0.5)
plt.axvline(x=0, color='white', linewidth=1, alpha=0.5)

# Grade
plt.grid(True, alpha=0.2)

# Vetores da base canônica
plt.quiver(0, 0, 0, -1, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='f = (0,-1)')
plt.quiver(0, 0, 1, 0, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='g = (1,0)')

# Pontos finais
plt.scatter([0, 1], [-1, 0], color=['red', 'blue'], s=50, zorder=5)

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend(fontsize=12)
plt.title('Base Canônica de ℝ²', fontsize=14)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

ax = plt.gca()  
ax.set_facecolor('black')  

plt.show()
```
