# Rotação horária em 90°:

 A transformação linear de rotação gira um vetor ao redor de um ponto fixo, num valor de ângulo específico, no caso da atividade em 90° horários. A matriz de rotação pode ser obtida seguindo duas lógicas distintas, a primeira é indo por lógica e verificando com vetores unitários o que ocorre quando eles são rotacionados, a partir deles pode ser montada uma matriz. A outra forma é por meio de formulação trigonométrica usando senθ e cosθ, mas que resulta na mesma matriz formada por um processo mais simples. 

#### Utilizando dois vetores unitários, da base canônica em ℝ², sendo eles:

$$
\vec{f}_ = \begin{pmatrix} 1 \\ 0 \end{pmatrix}
$$

$$
\vec{g}_ = \begin{pmatrix} 0 \\ 1 \end{pmatrix}
$$

```python
plt.figure(figsize=(8, 8))

# Plano cartesiano
plt.axhline(y=0, color='black', linewidth=1, alpha=0.5)
plt.axvline(x=0, color='black', linewidth=1, alpha=0.5)

# Grade
plt.grid(True, alpha=0.2)

# Vetores unitários(base canonica)
plt.quiver(0, 0, 1, 0, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='f = (1,0)')
plt.quiver(0, 0, 0, 1, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='g = (0,1)')

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend(fontsize=12)
plt.title('Vetores Unitários', fontsize=14)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')


plt.show()
```

Se aplica a rotação de 90°, transformando esses vetores resultando nos valores abaixo:

$$
\vec{f} = \begin{pmatrix} 0 \\ -1 \end{pmatrix}
$$


$$
\vec{g} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}
$$

```python
plt.figure(figsize=(8, 8))

# Plano cartesiano
plt.axhline(y=0, color='black', linewidth=1, alpha=0.5)
plt.axvline(x=0, color='black', linewidth=1, alpha=0.5)

# Grade
plt.grid(True, alpha=0.2)

# Vetores da base canônica
plt.quiver(0, 0, 0, -1, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='f = (0,-1)')
plt.quiver(0, 0, 1, 0, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='g = (1,0)')

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend(fontsize=12)
plt.title('Vetores Unitários Rotacionados', fontsize=14)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')



plt.show()
```

A partir dos vetores **g** e **f** rotacionados, se pode montar a matriz de rotação(matriz de transformação), transformado os vetores em colunas de uma matriz 2x2. Que gera a matriz $R_θ$, que é em suma a transformação linear representada de forma matricial. 

$$
R_θ = \begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
$$

Com a obtenção da matriz de transformação pode-se obter o valor transformado de qualquer vetor que seja multiplicado por ela, da seguinte forma: 

$$
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
\cdot
\begin{pmatrix}
x \\
y
\end{pmatrix}
=
\begin{pmatrix}
0x + 1y \\
-1x + 0y
\end{pmatrix}
$$

## Exemplo de Rotação em vetores quaisquer:

Dado dois vetores $k$ e $l$, eles serão rotacionados utilzando a matriz de transformação $R_θ$:

$$
\vec{k} = \begin{pmatrix} 4 \\ 2 \end{pmatrix}
$$

$$
\vec{l} = \begin{pmatrix} 3 \\ 1 \end{pmatrix}
$$


Multiplicação dos vetores $\vec{k}$ e $\vec{l}$ pela matriz de rotação de rotação $R_θ$:

Começando pelo vetor $\vec{k}$, que será multiplicado pela matriz e originará o vetor rotacionado: 

$$
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
\cdot
\begin{pmatrix}
4 \\
2
\end{pmatrix}
=
\begin{pmatrix}
0\cdot 4 + 1\cdot 2 \\
-1\cdot 4 + 0\cdot 2
\end{pmatrix}
=
\begin{pmatrix}
2 \\
-4
\end{pmatrix}
$$


Para o vetor $\vec{l}$, o mesmo, a fim de originar a rotação desse vetor: 

$$
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
\cdot
\begin{pmatrix}
3 \\
1
\end{pmatrix}
=
\begin{pmatrix}
0\cdot 3 + 1\cdot 1 \\
-1\cdot 3 + 0\cdot 1
\end{pmatrix}
=
\begin{pmatrix}
1 \\
-3
\end{pmatrix}
$$



### Por meio da multiplicação de cada vetor pela matriz obtem-se os gráficos:

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Graf 1
ax1.axhline(y=0, color='black', linewidth=1, alpha=0.5)
ax1.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax1.grid(True, alpha=0.2)

# Vetores i = (4,2) e j = (1,3)
ax1.quiver(0, 0, 4, 2, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='k = (4,2)')
ax1.quiver(0, 0, 3, 1, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='l = (3,1)')

ax1.set_xlim(-0.5, 5)
ax1.set_ylim(-0.5, 4)
ax1.legend(fontsize=10)
ax1.set_title('Vetores Antes da Rotação', fontsize=12)
ax1.set_xlabel('Eixo X')
ax1.set_ylabel('Eixo Y')

# graf 2
ax2.axhline(y=0, color='black', linewidth=1, alpha=0.5)
ax2.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax2.grid(True, alpha=0.2)

# Vetores (2,-4) e (1,-3)
ax2.quiver(0, 0, 2, -4, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='k = (2,-4)')
ax2.quiver(0, 0, 1, -3, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='l = (1,-3)')

ax2.set_xlim(-0.5, 3)
ax2.set_ylim(-5, 0.5)
ax2.legend(fontsize=10)
ax2.set_title('Vetores Rotacionados', fontsize=12)
ax2.set_xlabel('Eixo X')
ax2.set_ylabel('Eixo Y')

plt.tight_layout()
plt.show()
```

Após a transformação de Rotação os vetores **k** e **l** obtiveram os seguintes valores:

$$
\vec{k} = \begin{pmatrix} 2 \\ -4 \end{pmatrix}
$$

$$
\vec{l} = \begin{pmatrix} 1 \\ -3 \end{pmatrix}
$$


# Cisalhamento(Shear) no eixo X, plano horizontal: 

#### Cisalhamento é uma transformação geométrica que inclina/distorce objetos, podendo ser horizontal e vertical. A distorção ocorre de maneira proporcional em razão de um fator $k$ de cisalhamento.

$$
T: \mathbb{R}^2 \to \mathbb{R}^2
$$
$$
T(x, y) = (x + ky, y)
$$

Dado os vetores unitários $m$ e $n$, de uma base canônica ℝ², com os seguintes valores:

$$
\vec{m} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}
$$

$$
\vec{n} = \begin{pmatrix}  0 \\ 1 \end{pmatrix}
$$

Dados os valores da base canônica, se define um valor para o fator $k$ de cisalhamento.
<br>
Definirei o fator de cisalhamento como $k=3$.
<br>
Após isso efetua-se a transformação linear de cisalhamento em ambos os vetores de forma separada.

$$
T_m(1, 0) = (1 + 3\cdot 0, 0)
$$

$$
T_n(0, 1) = (0 + 3\cdot 1, 1)
$$

Que vai resultar nos seguintes valores transformados:

$$
\vec{m} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}
$$

$$
\vec{n} = \begin{pmatrix}  3 \\ 1 \end{pmatrix}
$$


```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Graf 1
ax1.axhline(y=0, color='black', linewidth=1, alpha=0.5)
ax1.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax1.grid(True, alpha=0.2)

# Vetores (1,0) e (0,1)
ax1.quiver(0, 0, 1, 0, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='(1,0)')
ax1.quiver(0, 0, 0, 1, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='(0,1)')

ax1.set_xlim(-1, 2)
ax1.set_ylim(-1, 2)
ax1.legend(fontsize=10)
ax1.set_title('Vetores na Base Canônica', fontsize=12)
ax1.set_xlabel('Eixo X')
ax1.set_ylabel('Eixo Y')

# Graf 2
ax2.axhline(y=0, color='black', linewidth=1, alpha=0.5)
ax2.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax2.grid(True, alpha=0.2)

# Vetores (1,0) e (3,1)
ax2.quiver(0, 0, 1, 0, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='(1,0)')
ax2.quiver(0, 0, 3, 1, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='(3,1)')

ax2.set_xlim(-0.5, 4)
ax2.set_ylim(-0.5, 2)
ax2.legend(fontsize=10)
ax2.set_title('Vetores após o Cisalhamento', fontsize=12)
ax2.set_xlabel('Eixo X')
ax2.set_ylabel('Eixo Y')

plt.tight_layout()
plt.show()
```

Com os novos valores dos vetores, se cria a matriz de transformação de cisalhamento $C_k$, juntando os vetores em colunas de uma matriz $2\times 2$.

$$
C_k = \begin{bmatrix}
1 & 3 \\
0 & 1
\end{bmatrix}
$$

Obtida a matriz de transformação $C_k$ já é possível aplicar vetores de valor quaisquer e observa-los transformados.


$$
\begin{bmatrix}
1 & 3 \\
0 & 1
\end{bmatrix}
\cdot
\begin{pmatrix}
x \\
y
\end{pmatrix}
=
\begin{pmatrix}
x + 3y \\
0x + y
\end{pmatrix}
$$

# Exemplo de Cisalhamento em vetores quaisquer: 
Dado os vetores $q$ e $w$, com os seguintes valores: 

$$
\vec{q} = \begin{pmatrix} 1 \\ 2 \end{pmatrix}
$$

$$
\vec{w} = \begin{pmatrix}  3 \\ 3 \end{pmatrix}
$$

Aplica-se a matriz de transformação nesses dois vetores:

$$
\begin{bmatrix}
1 & 3 \\
0 & 1
\end{bmatrix}
\cdot
\begin{pmatrix}
1 \\
2
\end{pmatrix}
=
\begin{pmatrix}
1 + 3\cdot 2 \\
0\cdot 1 + 2
\end{pmatrix}
=
\begin{pmatrix}
7 \\
2
\end{pmatrix}
$$



$$
\begin{bmatrix}
1 & 3 \\
0 & 1
\end{bmatrix}
\cdot
\begin{pmatrix}
3 \\
3
\end{pmatrix}
=
\begin{pmatrix}
3 + 3\cdot 3 \\
0\cdot 3 + 3
\end{pmatrix}
=
\begin{pmatrix}
12 \\
3
\end{pmatrix}
$$

Que vai resultar nos seguintes vetores: 

$$
\vec{q} = \begin{pmatrix} 7 \\ 2 \end{pmatrix}
$$

$$
\vec{w} = \begin{pmatrix}  12 \\ 3 \end{pmatrix}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Graf 1
ax1.axhline(y=0, color='black', linewidth=1, alpha=0.5)
ax1.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax1.grid(True, alpha=0.2)

# Vetores (1,2) e (3,3)
ax1.quiver(0, 0, 1, 2, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='(1,2)') #width é pra calibrar a largura da seta #label serve só para informar
ax1.quiver(0, 0, 3, 3, color='blue', scale=1, scale_units='xy', #color é para cor das setas
           angles='xy', width=0.015, label='(3,3)')

ax1.set_xlim(-1, 5)
ax1.set_ylim(-1, 5)
ax1.legend(fontsize=10)
ax1.set_title('Vetores Exemplo', fontsize=12)
ax1.set_xlabel('Eixo X') #nome no eixo X
ax1.set_ylabel('Eixo Y') #nome no eixo Y

# Graf 2
ax2.axhline(y=0, color='black', linewidth=1, alpha=0.5) 
ax2.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax2.grid(True, alpha=0.2)



# Vetores (7,2) e (12,3)
ax2.quiver(0, 0, 7, 2, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='(7,2)')
ax2.quiver(0, 0, 12, 3, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='(12,3)')

ax2.set_xlim(-0.5, 13)
ax2.set_ylim(-0.5, 4)

ax2.set_xticks(np.arange(0, 15, 1))  # começa no 0 e vai até antes do número do meio

ax2.legend(fontsize=10)
ax2.set_title('Vetores após o Cisalhamento', fontsize=12)
ax2.set_xlabel('Eixo X')
ax2.set_ylabel('Eixo Y')

plt.tight_layout()
plt.show()
```

### Exemplo de vetores com ambas as transformações:

Dados dois vetores, $t_1$ e $t_2$ de valores:

$$
\vec{t_1} = \begin{pmatrix} 2 \\ 3 \end{pmatrix}
$$

$$
\vec{t_2} = \begin{pmatrix}  2 \\ 3 \end{pmatrix}
$$

- O vetor $\vec{t_1}$ terá primeiro uma rotação e depois o cisalhamento.
<br>

- O vetor $\vec{t_2}$ terá primeiro um cisalhamento e depois uma rotação.

Começando em $\vec{t_1}$:
<br>
Aplicando a matriz de Rotação no vetor:

$$
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
\cdot
\begin{pmatrix}
2 \\
3
\end{pmatrix}
=
\begin{pmatrix}
0\cdot 2 + 1\cdot 3 \\
-1\cdot 2 + 0\cdot 3
\end{pmatrix}
$$

Que gera o vetor: 

$$
\vec{t_1} = \begin{pmatrix} 3 \\ -2 \end{pmatrix}
$$

Aplicando Cisalhamento: 

$$
\begin{bmatrix}
1 & 3 \\
0 & 1
\end{bmatrix}
\cdot
\begin{pmatrix}
3 \\
-2
\end{pmatrix}
=
\begin{pmatrix}
3 + 3\cdot -2 \\
0\cdot 3 + -2
\end{pmatrix}
$$

Resultando no vetor $\vec{t_1}$ com valor: 

$$
\vec{t_1} = \begin{pmatrix} -3 \\ -2 \end{pmatrix}
$$

Esse vetor sofreu Rotação e Cisalhamento respectivamente.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1) = plt.subplots(1, figsize=(14, 7))

ax1.axhline(y=0, color='black', linewidth=1, alpha=0.5)
ax1.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax1.grid(True, alpha=0.2)

# Vetores (2,3) e (3,-2) e (-3,-2)
ax1.quiver(0, 0, 2, 3, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.005, label='(2,3)N')
ax1.quiver(0, 0, 3, -2, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.005, label='(3,-2)R')
ax1.quiver(0, 0, -3, -2, color='green', scale=1, scale_units='xy', 
           angles='xy', width=0.005, label='(-3,-2)RC')

ax1.set_xlim(-5, 5)
ax1.set_ylim(-5, 5)

ax1.set_xticks(np.arange(-6, 6, 1))
ax1.set_yticks(np.arange(-6, 6, 1))

ax1.legend(fontsize=10)
ax1.set_title('Vetor Normal(N), Rotacionado(R), Rotacionado e Cisalhado(RC)', fontsize=12)
ax1.set_xlabel('Eixo X')
ax1.set_ylabel('Eixo Y')

plt.tight_layout()
plt.show()
```

Começando com $\vec{t_2}$:
<br>
Aplicando Cisalhamento:

$$
\begin{bmatrix}
1 & 3 \\
0 & 1
\end{bmatrix}
\cdot
\begin{pmatrix}
2 \\
3
\end{pmatrix}
=
\begin{pmatrix}
2 + 3\cdot 3 \\
0\cdot 2 + 3
\end{pmatrix}
$$

Que gera o vetor: 

$$
\vec{t_2} = \begin{pmatrix}  11 \\ 3 \end{pmatrix}
$$

Aplicando a Rotação:

$$
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
\cdot
\begin{pmatrix}
11 \\
3
\end{pmatrix}
=
\begin{pmatrix}
0\cdot 11 + 1\cdot 3 \\
-1\cdot 11 + 0\cdot 3
\end{pmatrix}
$$

Resultando no vetor $\vec{t_2}$ com valor: 

$$
\vec{t_2} = \begin{pmatrix} 3 \\ -11 \end{pmatrix}
$$

Esse vetor sofreu Cisalhamento e Rotação, respectivamente.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1) = plt.subplots(1, figsize=(14, 7))

ax1.axhline(y=0, color='black', linewidth=1, alpha=0.5)
ax1.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax1.grid(True, alpha=0.2)

# Vetores (2,3) e (11,3) e (3,-11)
ax1.quiver(0, 0, 2, 3, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.005, label='(2,3)N')
ax1.quiver(0, 0, 11, 3, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.005, label='(11,3)C')
ax1.quiver(0, 0, 3, -11, color='green', scale=1, scale_units='xy', 
           angles='xy', width=0.005, label='(3,-11)CR')

ax1.set_xlim(-12, 12)
ax1.set_ylim(-12, 12)

ax1.set_xticks(np.arange(-13, 13, 1))
ax1.set_yticks(np.arange(-13, 13, 1))

ax1.legend(fontsize=10)
ax1.set_title('Vetor Normal(N), Cisalhado(C), Cisalhado e Rotacionado(CR)', fontsize=12)
ax1.set_xlabel('Eixo X')
ax1.set_ylabel('Eixo Y')

plt.tight_layout()
plt.show()
```
