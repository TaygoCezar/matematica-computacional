##### A transformação linear de rotação gira um vetor ao redor de um ponto fixo, num valor de ângulo específico, no caso da atividade em 90° horários. A matriz de rotação pode ser obtida seguindo duas lógicas distintas, a primeira é indo por lógica e verificando com vetores unitários o que ocorre quando eles são rotacionados, a partir deles pode ser montada uma matriz. A outra forma é por meio de formulação trigonométrica usando senθ e cosθ, mas que resulta na mesma matriz formada por um processo mais simples. 

### Rotação horária em 90°:
Seguindo pela lógica menos complicada, utilizando dois vetores unitários, da base canônica em ℝ², sendo eles:

$$
\vec{f}_ = \begin{pmatrix} 1 \\ 0 \end{pmatrix}
$$

$$
\vec{g}_ = \begin{pmatrix} 0 \\ 1 \end{pmatrix}
$$

$$
R = \begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
$$

```python
plt.figure(figsize=(8, 8))

# Plano cartesiano
plt.axhline(y=0, color='white', linewidth=1, alpha=0.5)
plt.axvline(x=0, color='white', linewidth=1, alpha=0.5)

# Grade
plt.grid(True, alpha=0.2)

# Vetores da base canônica
plt.quiver(0, 0, 1, 0, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='f = (1,0)')
plt.quiver(0, 0, 0, 1, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='g = (0,1)')

# Pontos finais
plt.scatter([1, 0], [0, 1], color=['red', 'blue'], s=50, zorder=5)

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend(fontsize=12)
plt.title('Base Canônica de ℝ²', fontsize=14)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

ax = plt.gca()  
ax.set_facecolor('black')  

plt.show()
```

Se aplica a rotação de 90°, transformando esses vetores em: [0,-1] e [1,0], resultando nos vetores abaixo:

```python
plt.figure(figsize=(8, 8))

# Plano cartesiano
plt.axhline(y=0, color='white', linewidth=1, alpha=0.5)
plt.axvline(x=0, color='white', linewidth=1, alpha=0.5)

# Grade
plt.grid(True, alpha=0.2)

# Vetores da base canônica
plt.quiver(0, 0, 0, -1, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='f = (0,-1)')
plt.quiver(0, 0, 1, 0, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='g = (1,0)')

# Pontos finais
plt.scatter([0, 1], [-1, 0], color=['red', 'blue'], s=50, zorder=5)

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend(fontsize=12)
plt.title('Base Canônica de ℝ²', fontsize=14)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

ax = plt.gca()  
ax.set_facecolor('black')  

plt.show()
```

A partir dos vetores **g** e **f** rotacionados, se pode montar a matriz de rotação(matriz de transformação), transformado os vetores em colunas de uma matriz 2x2. Que gera a matriz $R_θ$, que é em suma a transformação linear representada de forma matricial. 

$$
R_θ = \begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
$$

Com a obtenção da matriz de transformação pode-se obter o valor transformado de qualquer vetor que seja multiplicado por ela, da seguinte forma: 

$$
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
\cdot
\begin{pmatrix}
x \\
y
\end{pmatrix}
=
\begin{pmatrix}
0x + 1y \\
-1x + 0y
\end{pmatrix}
$$

## Exemplo de Rotação em vetores quaisquer:

Dado dois vetores $k$ e $l$, eles serão rotacionados utilzando a matriz de transformação $R_θ$:

$$
\vec{k}_ = \begin{pmatrix} 4 \\ 2 \end{pmatrix}
$$

$$
\vec{l}_ = \begin{pmatrix} 3 \\ 1 \end{pmatrix}
$$

### Por meio da multiplicação de cada vetor pela matriz obtem-se os gráficos:


```python
import matplotlib.pyplot as plt
import numpy as np

# Criar figura com 1 linha e 2 colunas
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# --- PRIMEIRO GRÁFICO ---
ax1.axhline(y=0, color='black', linewidth=1, alpha=0.5)
ax1.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax1.grid(True, alpha=0.2)

# Vetores i = (4,2) e j = (1,3)
ax1.quiver(0, 0, 4, 2, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='i = (4,2)')
ax1.quiver(0, 0, 1, 3, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='j = (1,3)')

ax1.scatter([4, 1], [2, 3], color=['red', 'blue'], s=50, zorder=5)
ax1.set_xlim(-0.5, 5)
ax1.set_ylim(-0.5, 4)
ax1.legend(fontsize=10)
ax1.set_title('Vetores Antes da Rotação', fontsize=12)
ax1.set_xlabel('Eixo X')
ax1.set_ylabel('Eixo Y')

# --- SEGUNDO GRÁFICO ---
ax2.axhline(y=0, color='black', linewidth=1, alpha=0.5)
ax2.axvline(x=0, color='black', linewidth=1, alpha=0.5)
ax2.grid(True, alpha=0.2)

# Vetores (2,-4) e (1,-3)
ax2.quiver(0, 0, 2, -4, color='red', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='(2,-4)')
ax2.quiver(0, 0, 1, -3, color='blue', scale=1, scale_units='xy', 
           angles='xy', width=0.015, label='(1,-3)')

ax2.scatter([2, 1], [-4, -3], color=['red', 'blue'], s=50, zorder=5)
ax2.set_xlim(-0.5, 3)
ax2.set_ylim(-5, 0.5)
ax2.legend(fontsize=10)
ax2.set_title('Vetores Rotacionados', fontsize=12)
ax2.set_xlabel('Eixo X')
ax2.set_ylabel('Eixo Y')

plt.tight_layout()
plt.show()
```

Após a transformação de Rotação os vetores **k** e **l** obtiveram os seguintes valores:

$$
\vec{k}_ = \begin{pmatrix} 2 \\ -4 \end{pmatrix}
$$

$$
\vec{l}_ = \begin{pmatrix} 1 \\ -3 \end{pmatrix}
$$
